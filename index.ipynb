```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import kurtosis, skew
```

## Take 1
* Generate a random normal variable `x_random` in numpy with 10,000 values. Set the mean value to 0 and standard deviation to 2.
* Plot a histogram for the data, set bins to `auto` (default). 
* Calculate the skewness and kurtosis for this data distribution using the SciPy functions.
* Record your observations about the calculated values and the shape of the data. 

```python
plt.style.use('seaborn')
```

```python
mu, sigma = 0, 2
n = 10000

x_random = np.random.normal(mu,sigma,n)

print("Skewness =", skew(x_random))
print("Kurtosis =", kurtosis(x_random))
sns.distplot(x_random)

# Skewness = -0.0025781248189666343
# kurtosis = 0.03976806960642154
```

```python
# This random dataset has a slight positive skewness, meaning there are more values on the left
# The excess kurtosis is < 0, showing a Platykurtic distribution (subtract 3 from Pearson kurtosis)
# However, the kurtosis is very close to zero, meaning the data are almost perfectly normal
```

## Take 2

Let's generate another distribution 

```python
x = np.linspace( -5, 5, 10000 )
y = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x)**2  )  # normal distribution
sns.distplot(y)
```

* Plot a histogram for data $y$, and set bins to auto (default).
* Calculate the skewness and kurtosis for this data distribution using the SciPy functions.
* Record your observations about the calculated values and the shape of the data.

```python
print("Skewness =", skew(y))
print("Kurtosis =", kurtosis(y))

# Skewness = 1.109511549276228
# kurtosis = -0.31039027765889804
```

```python
# Your observations here 

# Very positive skew in data, showing the left side has more values than the right
# Slightly negative kurtosis, which would normally mean a Platykurtic distribution with a
# broader peak and thinner tails, but in this case since the data is obviously not normally
# distributed, the kurtosis value here is misleading and meaningless.
```
